```python
plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(data["Insulin"], data["Glucose"], c="red", alpha=0.3)
plt.ylabel("Glucose")
plt.xlabel("Insulin")
plt.subplot(2,1,2)
plt.scatter(data["BloodPressure"], data["Glucose"], c="blue", alpha=0.1)
plt.xlabel("Blood Pressure")
plt.ylabel("Glucose")
plt.show()

#Glucose and Insulin scatter plot is heavily distributed in the 0 to 200 range. We can infer that as the glucose level increases, amount of insulin 
#taken also increases. Glucose levels of those who take insulin within the range 0 to 200 are between 50 to 150 on average. A positive correlation can 
#be seen between gluose and insulin, because as the glucose levels increase, amount of insulin consumed also increases.
```
